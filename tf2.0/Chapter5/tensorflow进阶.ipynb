## 5.1 合并与分割
### 5.1.1 合并
张量的合并可以使用拼接和堆叠操作实现。拼接操作并不会产生新的维度，仅在现有的维度上合并，而堆叠操作会创建新维度。

```python
import tensorflow as tf
from tensorflow import keras
```

```python
a = tf.random.normal([4, 35, 8])
b = tf.random.normal([6, 35, 8])
tf.concat([a, b], axis=0) # 拼接合并，按照第一个维度进行合并
```

```python
a = tf.random.normal([35, 8])
b = tf.random.normal([35, 8])
tf.stack([a, b], axis=0) # 堆叠合并成2个班级，班级维度插在最前面
```

### 5.1.2 分割
通过tf.split(x, num_or_size_splits, axis)可以完成张量的分割操作，参数意义为：

x : 待分割张量

num_or_size_splits : 分割方案，为单个数值时，如10，表示等长分割成10份；当为List时，如[2, 4, 2, 2]表示分割成4份，每份长度分别为2，4，2，2

axis : 指定分割的维度索引号

```python
x = tf.random.normal([10, 35, 8])
result = tf.split(x, num_or_size_splits=10, axis=0)
len(result)
```

## 5.2 数据统计
### 5.2.1 向量范数
向量范数(Vector Norm)是表征向量“长度”的一种度量方法，他可以推广到张量上。常用的向量范数
* L1范数 : 定义为向量x的所有元素绝对值之和 $ \| x \| _1 = \sum_i{|x_i|} $
* L2范数 : 定义为向量x的所有元素的平方和，再开根号 $ \| x \|_2 = \sqrt{\sum_i |x_i| ^ 2} $
* ∞-范数 : 定义为向量x的所有元素绝对值的最大值 $ \| x \| _ \infty = \max_i \left( | x_i | \right) $

在TensorFlow中，可以通过tf.norm(x, ord)来求解张量的L1、L2、∞等范数，其中指定为1、2时计算L1、L2范数，指定为np.inf时计算∞-范数

```python
x = tf.ones([2, 2])
tf.norm(x, ord=1)
```

```python
tf.norm(x, ord=2)
```

```python
import numpy as np
tf.norm(x, ord=np.inf)
```

### 5.2.2 最值、均值、和
通过tf.reduce_max、tf.reduce_min、tf.reduce_mean、tf.reduce_sum函数可以求解张量在某个维度上的最大、最小、均值、和，也可以求解全局最大、最小、均值、和

```python
x = tf.random.normal([4, 10])
tf.reduce_max(x, axis=1)
```

```python
# 不指定axis参数时，会计算全局的目标值
tf.reduce_max(x) # 返回结果为标量
```

在求解误差函数时，通过TensorFLow的MSE误差函数可以求得每个样本得误差，需要计算样本的误差，此时可以通过tf.reduce_mean在样本维度上计算均值

```python
out = tf.random.normal([4, 10]) # 模拟网络预测输出
y = tf.constant([1, 2, 2, 0]) # 模拟真实标签
y = tf.one_hot(y, depth=10)
loss = keras.losses.MSE(y, out) # 计算每个样本的误差
loss = tf.reduce_mean(loss) # 平均误差
loss
```

除了希望获取张量的最值信息，还希望获得最值所在的位置索引号，例如分类任务的标签预测，就需要知道**概率最大值**所在的索引号，一般把这个位置索引号作为预测类别。

```python
out = tf.random.normal([2, 10])
out = tf.nn.softmax(out, axis=1) # 通过softmax函数转换为概率值
out
```

通过tf.argmax(x, axis)和tf.argmin(x, axis)可以求解在axis维度上，x的最大值、最小值所在的索引号

```python
pred = tf.argmax(out, axis=1) # 选择概率最大的位置
pred
```

## 5.3 张量比较
为了计算分类任务的准确率等指标，一般需要将预测结果和真实标签比较，统计比较结果中正确的数据来计算准确率。

通过tf.softmax获取预测类别

```python
out = tf.random.normal([100, 10])
out = tf.nn.softmax(out, axis=1) # 输出转换为概率
pred = tf.argmax(out, axis=1) # 计算预测值
pred
```

```python
# 模型生成真实标签
y = tf.random.uniform([100], dtype=tf.int64, maxval=10)
out = tf.equal(pred, y) # 预测值与真实值比较，返回布尔型的张量
out
```

```python
out = tf.cast(out, dtype=tf.float32) # 布尔型转int型
correct = tf.reduce_sum(out) # 统计True的个数
correct.numpy()
```

## 5.4 填充与复制
### 5.4.1 填充
对于图片数据的高和宽、序列信号的长度，维度信号长度可能各不相同。为了方便网络的并行计算，需要将不同长度的数据扩展为相同长度

通过tf.pad(x, paddings)函数实现，参数paddings是包含了多个[Left Padding, Right Padding]的嵌套方案List，如[[0, 0], [2, 1], [1, 2]]表示第一个维度不变，第二个维度左边（起始边）填充两单元，右边（结束端）填充一单元，第三个维度左边（起始边）填充一单元，右边（结束端）填充两单元。

```python
a = tf.constant([1, 2, 3, 4, 5, 6]) # 第一个句子
b = tf.constant([7, 8, 1, 6]) # 第二个句子
b = tf.pad(b, [[0, 2]]) # 句子末尾填充2个0
b
```

```python
# 填充后句子张量形状一致，可以Stack在一起
tf.stack([a, b], axis=0)
```

在自然语言处理中，需要加载不同句子长度的数据集，有些句子长度较小，如仅10个单词，部分句子较长，如超过100个单词。为了能够保存在同一张量中，一般会选取能够覆盖大部分句子长度的阈值，如80个单词的句子，截断超过规定长度的部分单词，以IMDB数据集的加载为例。

```python
total_words = 10000 # 设定词汇量大小
max_review_len = 80 # 最大句子长度
embedding_leb = 100  # 词向量长度
# 加载IMDB数据集
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.imdb.load_data(num_words=total_words)
# 将句子填充或截断到相同长度，设置为末尾填充和末尾截断方式
x_train = keras.preprocessing.sequence.pad_sequences(x_train, maxlen=max_review_len, truncating='post', padding='post')
x_test = keras.preprocessing.sequence.pad_sequences(x_test, maxlen=max_review_len, truncating='post', padding='post')
print(x_train.shape, x_test.shape)
```

## 5.6 高级操作
### 5.6.1 tf.gather
通过tf.gather函数可以收集（分割）数据，相对于切片分割而言，更适合于**不规则**的切片规则

同时tf.gather函数中的索引号可以乱序排列，此时收集的数据也是对应的乱序顺序

```python
x = tf.random.uniform([4, 35, 8], maxval=100, dtype=tf.int32)
tf.gather(x, [0, 3, 8, 11, 12, 26], axis=1)
```

```python
tf.gather(x, [4, 2], axis=2)
```

### 5.6.2 tf.gather_nd
通过tf.gather_nd函数，可以通过指定每次采样点的多维坐标来实现采样多个点的目的。

```python
tf.gather_nd(x, [[1, 1], [2, 2], [3, 3]]) # 根据多维坐标收集数据
```

### 5.6.3 tf.boolean_mask
通过掩码(mask)的方式进行采样。

掩码List长度必须与给定维度的长度相同

```python
# 根据掩码方式采样班级，给出掩码和维度索引
tf.boolean_mask(x, mask=[True, False, False, True], axis=0)
```

### 5.6.4 tf.where
通过tf.where(cond, a, b)操作可以根据cond条件的真假从参数A或B中读取数据

```python
a = tf.ones([3, 3])
b = tf.zeros([3, 3])
# 构造采样条件
cond = tf.constant([[True, False, False], [False, True, False], [True, True, False]])
tf.where(cond, a, b)
```

通过比较得到所有正数的掩码，通过tf.where提取此掩码处True元素的索引坐标，通过tf.gather_nd即可恢复出所有正数的元素

```python
x = tf.random.normal([3, 3])
mask = x > 0
indices = tf.where(mask)
tf.gather_nd(x, indices)
```

### 5.6.5 tf.scatter_nd
通过tf.scatter_nd(indices, updates, shape)函数可以高效地刷新张量的部分数据，但是这个函数只能在全0的白板张量上面执行刷新操作

```python
# 构造需要刷新数据的位置
indices = tf.constant([[4], [3], [1], [7]])
# 构造需要写入的数据
updates = tf.constant([4.4, 3.3, 1.1, 7.7])
# 在长度为8的全0向量上根据indices写入updates数据
tf.scatter_nd(indices, updates, [8])
```

### 5.6.6 tf.meshgrid
通过tf.meshgrid函数可以方便地生成二维网络的采样点坐标，方便可视化等应用场合

```python
x = tf.linspace(-8., 8, 100)
y = tf.linspace(-8., 8, 100)
x, y = tf.meshgrid(x, y) # 生成网格点，并内部拆分后返回
print(x.shape)
print(y.shape)
```

```python
z = tf.sqrt(x**2 + y**2)
z = tf.sin(z) / z
```

```python
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)
ax.contour3D(x.numpy(), y.numpy(), z.numpy(), 50)
plt.show()
```
